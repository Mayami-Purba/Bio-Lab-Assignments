# ba1j

```python
from collections import defaultdict

def HammingDistance(s1, s2):
    d=0
    for i in range(len(s1)):
        if s1[i]!=s2[i]:
            d += 1
    
    return d

def ReversePattern(dna):
    complement = {'A':'T' , 'C':'G' , 'T':'A' , 'G':'C'}
    rc=""
    for letter in dna:
        rc += complement[letter]
    
    rc = reversed(rc)  
    return ("".join(rc))

def new(pattern, mis, words):
    if mis == 0:
        words.add(pattern)
    else:
        letter = ['A', 'T', 'C', 'G']
        for i in range(len(pattern)):
            for j in range(len(letter)):
                np = pattern[:i] + letter[j] + pattern[i+1:]
                if mis <= 1:
                    words.add(np)
                else:
                    new(np, mis-1, words)

def FindMostFrequentPattern(text, k, d):
    allwords = defaultdict(int)
    for i in range(len(text) - k + 1):
        freq = set()
        new(text[i:i + k], d, freq)

        for w in freq:
            allwords[w] += 1

    for t in allwords.keys():
        rev = ReversePattern(t)
        for i in range(len(text) - k + 1):
            if HammingDistance(text[i:i + k], rev) <= d:
                allwords[t] += 1

    result = set()
    for t in allwords.keys():
        if allwords[t] == max(allwords.values()):
            result.add(t)
            result.add(ReversePattern(t))
    for i in result:
        print(i, end=" ")
        


text = input("Enter the text: ")
k = int(input("Enter the k-value: "))
d = int(input("Enter the hamming distance: "))        
FindMostFrequentPattern(text, k, d)
```

```python
def kmer(k):
    base = ['A', 'C', 'G', 'T']
    arr = base
    for n in range(k-1):
        arr = [i+j for i in arr for j in base]
        
    return arr

def PatternCount(text, pattern):
    count = 0
    for i in range(len(text)-len(pattern)+1):
        if text[i:i+len(pattern)] == pattern:
            count += 1
    return count

text = input()
k = int(input())
arr = kmer(k)

for pt in arr:
        print(PatternCount(text, pt), end=" ")

#ACGCGGCTCTGAAA
#2
```

# BA1N - Generate the d-Neighborhood of a String

```python
dna = input()
k = int(input())
dlen = len(dna)
arr = kmer(dlen)
nb = []

for ar in arr:
    if HammingDistance(dna, ar) <= k:
        nb.append(ar)
        
for i in range(len(nb)):
    print(''.join(nb[i]))
```


# BA2A - Implement MotifEnumeration

```python
def Motif(string, mis, words):
    if mis == 0:
        words.add(string)
    else:
        bases = ['A', 'T', 'C', 'G']
        
        for i in range(len(string)):
            for j in range(len(bases)):
                new = string[:i] + bases[j] + string[i+1:]
                if mis <= 1:
                    words.add(new)
                else:
                    motif(new, mis-1, words)

def MotifEnumeration(dna, k, d):
    pt1 = []
    dlen = len(dna)
    
    for n in range(dlen):
        pt2 = set()
        for i in range(len(dna[n]) - k + 1):
            pattern = set()
            Motif(dna[n][i:i + k], d,pattern)
            for kmer in pattern:
                pt2.add(kmer)
                
        for j in pt2:
            pt1.append(j)
            
    motif = []
    
    for pt in pt1:
        if pt1.count(pt) == dlen:
            motif.append(pt)
            
    motif = list(set(motif))
    return motif

k = int(input())
d = int(input())
dna = []
a = ''

while True:
    a = input()
    if a == 'quit':
        break
    dna.append(a)
    
ans = MotifEnumeration(dna, k, d)
print(" ".join(ans))

#3 1
#ATTTGGC
#TGCCTTA
#CGGTATC
#GAAAATT
    
```

# BA2B - Find a Median String

```python
def GenPattern(k):
    bases = ['A', 'C', 'G', 'T']
    array = bases
    for n in range(k-1):
        array = [i+j for i in array for j in bases]
    return array

def MinHammingDistance(pattern, string):
    min_distance = len(pattern)
    for i in range(len(string) - len(pattern) + 1):
        distance = sum([1 for j in range(len(pattern)) if pattern[j] != string[i:i+len(pattern)][j]])
        if distance < min_distance:
            min_distance = distance
    return min_distance


k = int(input())

dna1 = input()
dna2 = input()
dna3 = input()
dna4 = input()
dna5 = input()

dna = [dna1,dna2,dna3,dna4,dna5]

pattern = GenPattern(k) 
dis1 = {}
min_string = len(dna) * len(pattern)

for i in pattern:
    dist = 0
    for j in range(len(dna)):
        dist += MinHammingDistance(i, dna[j])
    dis1[i] = dist
    if dist < min_string:
        min_string = dist
p=0;

for t in dis1.keys():
    if dis1[t] == min_string:
        p += 1
        print(t)
    if p == 1:
        break
        


#3
#AAATTGACGCAT
#GACGACCACGTT
#CGTCAGCGCCTG
#GCTGAGCACCGG
#AGTACGGGACAG
```

# BA2C - Find a Profile-most Probable k-mer in a String

# BA2D - Implement GreedyMotifSearch

```python
def Probability(string, matrix):
    count = 1
    for i in range(len(string)):
        if string[i] == 'A':
            count *= matrix[0][i]
        if string[i] == 'C':
            count *= matrix[1][i]
        if string[i] == 'G':
            count *= matrix[2][i]
        if string[i] == 'T':
            count *= matrix[3][i]
    return count

def MostProbable(string, k, matrix):
    lt = {}
    for i in range(len(string)-k+1):
        lt[string[i:i+k]] = Probability(string[i:i+k], matrix)
    most = sorted(lt.items(), key=lambda x:x[1], reverse=True)[0][0]
    
    return most


def Score(motifs):
    s = 0
    for i in range(len(motifs[0])):
        j = []
        for mt in motifs:
            j.append(mt[i])
        s += (len(j) - max(j.count("A"), j.count("C"), j.count("T"), j.count("G")))
    return s


def GreedyMotifSearch(Dna, k, t):

    BestMotifs = []
    for dna in Dna:
        BestMotifs.append(dna[:k])
    
    Motifs = []
    
    for i in range(len(Dna[0])-k+1):
        for kmer in [Dna[0][i : i+k]]:
            NewMotifs = [kmer] 
            for i in range(1, t):          
                motifs = NewMotifs[:i]
                matrix = []
                for lt in ["A", "C", "G", "T"]:
                    mat = []
                    for j in range(k):
                        mm = []
                        for mt in motifs:
                            mm.append(mt[j])
                        mat.append(mm.count(lt)/len(motifs))
                    matrix.append(mat)
                            
                NewMotifs.append(MostProbable(Dna[i], k, matrix))
                
            if Score(NewMotifs) < Score(BestMotifs):               
                BestMotifs = NewMotifs
    
    return BestMotifs



k = int(input())
t = int(input())
Dna = []

for i in range(t):
    a = input()
    Dna.append(a)
    
BestMotifs = GreedyMotifSearch(Dna, k ,t)

for i in range(len(BestMotifs)):
    print("".join(BestMotifs[i]))


#3
#5
#GGCGTTCAGGCA
#AAGAATCAGTCA
#CAAGGAGTTCGC
#CACGTCAATCAC
#CAATAATATTCG
```

# BA2E - Implement GreedyMotifSearch with Pseudocounts

```python
def GreedyMotifSearchPsedo(Dna, k, t):

    BestMotifs = []
    for dna in Dna:
        BestMotifs.append(dna[:k])
    
    Motifs = []
    
    for i in range(len(Dna[0])-k+1):
        for kmer in [Dna[0][i : i+k]]:
            NewMotifs = [kmer] 
            for i in range(1, t):          
                motifs = NewMotifs[:i]
                matrix = []
                for lt in ["A", "C", "G", "T"]:
                    mat = []
                    for j in range(k):
                        mm = []
                        for mt in motifs:
                            mm.append(mt[j])
                        mat.append(mm.count(lt)+1/(2*len(motifs)))
                    matrix.append(mat)
                            
                NewMotifs.append(MostProbable(Dna[i], k, matrix))
                
            if Score(NewMotifs) < Score(BestMotifs):               
                BestMotifs = NewMotifs
    
    return BestMotifs

k = int(input())
t = int(input())
Dna = []

for i in range(t):
    a = input()
    Dna.append(a)
    
BestMotifs = GreedyMotifSearchPsedo(Dna, k ,t)

for i in range(len(BestMotifs)):
    print("".join(BestMotifs[i]))


#3 5
#GGCGTTCAGGCA
#AAGAATCAGTCA
#CAAGGAGTTCGC
#CACGTCAATCAC
#CAATAATATTCG
```

# BA2F - Implement RandomizedMotifSearch

```python
import random
N = 1000

def RandomizeMotifSearch(Dna,k,t):
    Motifs = []

    for i in range(t):
        n = len(Dna[i])
        x = random.randint(0, n-k)
        Motifs.append(Dna[i][x:x+k])
    
    bestMotifs = Motifs
    
    while True:
        profile = []
        
        letter = ['A' , 'C' ,'G','T']
        
        for ch in letter:
            pr = []
            for i in range(k):
                prof = []
                for mt in Motifs:
                    prof.append(mt[i])                    
                result = (prof.count(ch)+1/(len(Motifs)*2))
                pr.append(result)
            profile.append(pr)
        
        for i in range(t):
            Motifs[i] = MostProbable(Dna[i], k, profile)
        
        if Score(Motifs) < Score(bestMotifs):
            bestMotifs = Motifs
            
        return bestMotifs
    

k = int(input())
t = int(input())
Dna = []

for i in range(t):
    a = input()
    Dna.append(a)
    
    
M = RandomizeMotifSearch(Dna, k, t)
bMotifs = M

for i in range(N):
    M = RandomizeMotifSearch(Dna, k, t)
    if Score(M) < Score(bMotifs):
         bMotifs = M
    else:
        bestMotifs = bMotifs

print ('\n'.join(bestMotifs))
#8 5
#CGCCCCTCTCGGGGGTGTTCAGTAAACGGCCA
#GGGCGAGGTATGTGTAAGTGCCAAGGTGCCAG
#TAGTACCGAGACCGAAAGAAGTATACAGGCGT
#TAGATCAAGTTTCAGGTGCACGTCGGTGAACC
#AATCCACCAGCTCCACGTGCAATGTTGGCCTA
```

# BA2G - Implement GibbsSampler

```python
import random
N = 20

def randomGenerateKmer(text,k,profile):
    x = len(text)-k+1
    kmer = []
    for i in range(x):
        kmer.append(text[i:i+k])
        
    #sum =0;
    pr = []
    for i in range(x):
        r = Probability(kmer[i],profile)
        pr.append(r)

    ind = index(pr,x)
    return kmer[ind]
    
def index(pr,x):
    random.seed()
    
    a = random.random() * sum(pr)
    sum_pr = 0;
    for i in range(x-1):
        if a>=sum_pr and a<sum_pr+pr[i]:
            return i
        sum_pr +=pr[i]
        
    return x-1

def gibbsSampling(Dna , k , t , n):
    Motifs = []

    for i in range(t):
        l = len(Dna[i])
        x = random.randint(0, l-k)
        Motifs.append(Dna[i][x:x+k])
    
    bestMotifs = Motifs
    c=0;
    
    while(c<n):
        c += 1
        j = random.randint(0,t-1)
        profile = []
        
        letter = ['A' , 'C' ,'G','T']
        
        for ch in letter:
            pr = []
            for i in range(k):
                prof = []
                for mt in Motifs:
                    prof.append(mt[:j]+mt[j+1:])                    
                result = (prof.count(ch)+1/len(Motifs)*2)
                pr.append(result)
            profile.append(pr) 
            
        r = randomGenerateKmer(Dna[j],k,profile)
        Motifs[j] = r
        
        #print(profile)
        #print(c)       
        
        if Score(Motifs) < Score(bestMotifs):
            bestMotifs = Motifs
                    
    return bestMotifs
        
    
k = int(input())
t = int(input())
n = int(input())
Dna = []

for i in range(t):
    a = input()
    Dna.append(a)
     
M = gibbsSampling(Dna, k, t,n)
bMotifs = M

for i in range(N):
    M = gibbsSampling(Dna, k, t,n)
    if Score(M) < Score(bMotifs):
         bMotifs = M
    else:
        bestMotifs = bMotifs


print ('\n'.join(bestMotifs))


#8 5 100
#CGCCCCTCTCGGGGGTGTTCAGTAAACGGCCA
#GGGCGAGGTATGTGTAAGTGCCAAGGTGCCAG
#TAGTACCGAGACCGAAAGAAGTATACAGGCGT
#TAGATCAAGTTTCAGGTGCACGTCGGTGAACC
#AATCCACCAGCTCCACGTGCAATGTTGGCCTA
```

# BA2H - Implement DistanceBetweenPatternAndStrings

```python
def HammingDistance(s1, s2):
    d=0
    for i in range(len(s1)):
        if s1[i]!=s2[i]:
            d += 1
    
    return d

pattern = input()
dna = []
a = ''
while True:
    a = input()
    if a == 'quit':
        break
    dna.append(a)
    
k = len(pattern)
dist = 0

for ss in dna:
    ham = float("inf")    
    for i in range(len(ss) - k + 1):
        res = HammingDistance(pattern, ss[i:i + k])
        if ham > res:
            ham = res
    dist = dist + ham
    
print(dist)

#AAA
#TTACCTTAAC GATATCTGTC ACGGCGTTCG CCCTAAAGAG CGTCAGAGGT
```

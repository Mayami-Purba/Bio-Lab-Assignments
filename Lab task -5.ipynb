```python
CodonTable = {
    'UUU': 'F',     'CUU': 'L',     'AUU': 'I',     'GUU': 'V',
    'UUC': 'F',     'CUC': 'L',     'AUC': 'I',     'GUC': 'V',
    'UUA': 'L',     'CUA': 'L',     'AUA': 'I',     'GUA': 'V',
    'UUG': 'L',     'CUG': 'L',     'AUG': 'M',     'GUG': 'V',
    'UCU': 'S',     'CCU': 'P',     'ACU': 'T',     'GCU': 'A',
    'UCC': 'S',     'CCC': 'P',     'ACC': 'T',     'GCC': 'A',
    'UCA': 'S',     'CCA': 'P',     'ACA': 'T',     'GCA': 'A',
    'UCG': 'S',     'CCG': 'P',     'ACG': 'T',     'GCG': 'A',
    'UAU': 'Y',     'CAU': 'H',     'AAU': 'N',     'GAU': 'D',
    'UAC': 'Y',     'CAC': 'H',     'AAC': 'N',     'GAC': 'D',
    'UAA': 'Stop',  'CAA': 'Q',     'AAA': 'K',     'GAA': 'E',
    'UAG': 'Stop',  'CAG': 'Q',     'AAG': 'K',     'GAG': 'E',
    'UGU': 'C',     'CGU': 'R',     'AGU': 'S',     'GGU': 'G',
    'UGC': 'C',     'CGC': 'R',     'AGC': 'S',     'GGC': 'G',
    'UGA': 'Stop',  'CGA': 'R',     'AGA': 'R',     'GGA': 'G',
    'UGG': 'W',     'CGG': 'R',     'AGG': 'R',     'GGG': 'G'
}

def revDna(dna):
    complement = {'A':'T' , 'C':'G' , 'T':'A' , 'G':'C'}   
    rc = ''

    for letter in dna:
        rc += complement[letter]
        
    rc = "".join(reversed(rc))
    return rc

def translation(rna):
    pr = ''
    
    for i in range(0,len(rna),3):
        s = CodonTable[rna[i:i+3]]
        if s == 'Stop':
            break
        pr += s
    
    return pr

def transcribe(dna):
    return dna.replace('T', 'U') 


dna = input("Enter the dna string: ")
peptide = input("Enter the amino acid string: ")

p = len(peptide)*3

s1 = []
s2 = []

for i in range(len(dna)-p+1):
    s1.append(dna[i:i+p])
    s2.append(revDna(dna[i:i+p]))

for i in range(len(s1)):
    t1 = transcribe(s1[i])
    t2 = transcribe(s2[i])
    r1 = translation(t1)
    r2 = translation(t2)
    
    if r1 == peptide or r2==peptide:
        print(s1[i])

```

```python
mass = {}
mass['G'] = 57
mass['A'] = 71
mass['S'] = 87
mass['P'] = 97
mass['V'] = 99
mass['T'] = 101
mass['C'] = 103
mass['I'] = 113
mass['L'] = 113
mass['N'] = 114
mass['D'] = 115
mass['K'] = 128
mass['Q'] = 128
mass['E'] = 129
mass['M'] = 131
mass['H'] = 137
mass['F'] = 147
mass['R'] = 156
mass['Y'] = 163
mass['W'] = 186


def total_Mass(peptide):
    total = 0
    for ch in peptide:
        total += mass[ch]
        
    return total

peptide = input("Enter the peptide string: ")

pl = len(peptide)
res = [0]

for x in range(1,pl):
    for i in range(pl):
        if i+x > len(peptide):
            y = i+x-len(peptide)
            res.append(total_Mass(peptide[i:]+peptide[:y]))
        else:
            res.append(total_Mass(peptide[i:i+x]))
            
res.append(total_Mass(peptide))
res.sort()

ans = ""

for x in res:
    ans +=  str(x) + " "
    
print(ans)
```

```python
def debruijn(patterns):
    kmers = []
    for pattern in patterns:
        kmers = kmers + composition(len(pattern), pattern, uniq=True)
        
    kmers = set(kmers)
    res = {}
    
    for k in kmers:
        res[k] = []
        
    for kmer in patterns:
        res[prefix(kmer)].append(suffix(kmer))
        
    return res


def GenomePath(reads):
    path = reads[0]
    for e in reads[1:]:
        path += e[len(e)-1:]

    return path

def eulerPath(dict):
    stack = []
    count = balance(dict)   
    stack.append([k for k, v in count.items() if v==-1][0])
    path = []
    
    while stack != []:
        u = stack[-1]        
        try:
            v = dict[u][0]
            stack.append(v)
            dict[u].remove(v)
        except:           
            path.append(stack.pop())
        
    return path[::-1]


def composition(k, text, uniq=False):
    kmers = []
    for i in range(len(text)+1-k):
        kmers.append(text[i:i+k-1])
    if uniq:
        return sorted(list(kmers))
    else:
        return sorted(kmers)

def balance(adj):
    count = dict.fromkeys(adj.keys(), 0)
    for node in adj.keys():
        for out in adj[node]:
            count[node] -= 1
            try:
                count[out] += 1
            except:
                count[out] = 1
    return count


def suffix(string):
    return string[1:]


def prefix(string):
    return string[0:-1]

k = int(input())
dna = []
a = ''
while True:
    a = input()
    if a == 'quit':
        break
    dna.append(a)

ans = GenomePath(eulerPath(debruijn(dna)))
print(ans)
    
#4
#CTTA
#ACCA
#TACC
#GGCT
#GCTT
#TTAC
```
